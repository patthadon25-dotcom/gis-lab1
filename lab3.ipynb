```python
import folium
import geopandas as gpd
import pandas as pd
```

```python
polbndry = gpd.read_file("Lab3_Dataset/polbndry.shp")
place = gpd.read_file("Lab3_Dataset/places.shp")
tambon = gpd.read_file("Lab3_Dataset/tambon.dbf",encoding='tis-620')
pl_type = gpd.read_file("Lab3_Dataset/PL_TYPE.dbf",encoding='tis-620')
```

```python
print(polbndry.crs)
print(place.crs)
```

```python
polbndry = polbndry.set_crs('EPSG:24047')
place = place.set_crs('EPSG:24047')
print(polbndry.crs)
print(place.crs)
```

```python
print(polbndry.columns)
print(tambon.columns)
print(place.columns)
print(pl_type.columns)
```

```python
polbndry = polbndry.merge(tambon, on = 'TAMBON_IDN', how = 'left')
place = place.merge(pl_type, on = 'PL_TYPE', how = 'left')
```

```python
polbndry.head()
```

```python
place.head()
```

```python
school = place[place['TYP_NAM_T'] == 'โรงเรียน'].copy()
school['geometry'] = school.buffer(5000)
schoolbuff5k = school.dissolve()
schoolbuff5k_clip = gpd.clip(schoolbuff5k, polbndry)
```

```python
m = folium.Map(location=[17.742408, 100.569862], zoom_start=9)
folium.GeoJson(schoolbuff5k_clip,name='schoolbuff5k').add_to(m)
folium.LayerControl().add_to(m)
m.save("map_lab3.html")
```
